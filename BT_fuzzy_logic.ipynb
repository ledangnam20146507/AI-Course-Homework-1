```python
from numpy.lib.function_base import average
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
food =ctrl.Antecedent(np.arange(0,11,1),'food')
service =ctrl.Antecedent(np.arange(0,11,1),'service')
tip = ctrl.Consequent(np.arange(0,31,1),'tip')
food['poor']=fuzz.trimf(food.universe,[0,0,5])
food['average']=fuzz.trimf(food.universe,[0,5,10])
food['good']=fuzz.trimf(food.universe,[5,10,10])
service['poor']=fuzz.trimf(service.universe,[0,0,5])
service['average']=fuzz.trimf(service.universe,[0,5,10])
service['good']=fuzz.trimf(service.universe,[5,10,10])
tip['good']=fuzz.trimf(tip.universe,[20,25,31])
tip['average']=fuzz.trimf(tip.universe,[10,15,20])
tip['poor']=fuzz.trimf(tip.universe,[0,6,10])
rule1=ctrl.Rule(food['poor']&service['poor'],tip['poor'])
rule2=ctrl.Rule(food['average']&service['poor'],tip['average'])
rule3=ctrl.Rule(food['good']&service['poor'],tip['average'])
rule4=ctrl.Rule(food['poor']&service['average'],tip['poor'])
rule5=ctrl.Rule(food['good']&service['average'],tip['good'])
rule6=ctrl.Rule(food['poor']&service['good'],tip['average'])
rule7=ctrl.Rule(food['average']&service['average'],tip['average'])
rule8=ctrl.Rule(food['average']&service['good'],tip['good'])
rule9=ctrl.Rule(food['good']&service['good'],tip['good'])
tipping_ctrl=ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9])
tipping=ctrl.ControlSystemSimulation(tipping_ctrl)
tipping.input['service']=9.8
tipping.input['food']=7.0
tipping.compute()
print(tipping.output['tip'])
tip.view(sim= tipping)
food.view()
service.view()
tip.view()




```

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
time =ctrl.Antecedent(np.arange(10,301,1),'time')
rice =ctrl.Antecedent(np.arange(100,2001,1),'rice')
power = ctrl.Consequent(np.arange(0,101,1),'power')
time['ratnhanh']=fuzz.trimf(time.universe,[10,15,20])
time['nhanh']=fuzz.trimf(time.universe,[20,35,40])
time['trungbinh']=fuzz.trimf(time.universe,[40,80,140])
time['cham']=fuzz.trimf(time.universe,[140,180,220])
time['ratcham']=fuzz.trimf(time.universe,[220,275,300])
time.view()
rice['ratnhe']=fuzz.trimf(rice.universe,[100,150,300])
rice['nhe']=fuzz.trimf(rice.universe,[300,320,450])
rice['TB']=fuzz.trimf(rice.universe,[450,800,1000])
rice['nang']=fuzz.trimf(rice.universe,[1000,1200,1500])
rice['ratnang']=fuzz.trimf(rice.universe,[1500,1750,2000])
rice.view()
power['ratmanh']=fuzz.trimf(power.universe,[90,97,100])
power['manh']=fuzz.trimf(power.universe,[65,75,90])
power['TB']=fuzz.trimf(power.universe,[50,56,65])
power['yeu']=fuzz.trimf(power.universe,[35,46,50])
power['ratyeu']=fuzz.trimf(power.universe,[10,24,33])
power.view()
rule1=ctrl.Rule(time['ratnhanh']&rice['ratnhe'],power['TB'])
rule2=ctrl.Rule(time['ratnhanh']&rice['nhe'],power['manh'])
rule3=ctrl.Rule(time['ratnhanh']&rice['TB'],power['manh'])
rule4=ctrl.Rule(time['ratnhanh']&rice['nang'],power['ratmanh'])
rule5=ctrl.Rule(time['ratnhanh']&rice['ratnang'],power['ratmanh'])
rule6=ctrl.Rule(time['nhanh']&rice['ratnhe'],power['TB'])
rule7=ctrl.Rule(time['nhanh']&rice['nhe'],power['TB'])
rule8=ctrl.Rule(time['nhanh']&rice['TB'],power['manh'])
rule9=ctrl.Rule(time['nhanh']&rice['nang'],power['manh'])
rule10=ctrl.Rule(time['nhanh']&rice['ratnang'],power['ratmanh'])
rule11=ctrl.Rule(time['trungbinh']&rice['ratnhe'],power['TB'])
rule12=ctrl.Rule(time['trungbinh']&rice['nhe'],power['TB'])
rule13=ctrl.Rule(time['trungbinh']&rice['TB'],power['manh'])
rule14=ctrl.Rule(time['trungbinh']&rice['nang'],power['manh'])
rule15=ctrl.Rule(time['trungbinh']&rice['ratnang'],power['manh'])
rule16=ctrl.Rule(time['cham']&rice['ratnhe'],power['yeu'])
rule17=ctrl.Rule(time['cham']&rice['nhe'],power['yeu'])
rule18=ctrl.Rule(time['cham']&rice['TB'],power['TB'])
rule19=ctrl.Rule(time['cham']&rice['nang'],power['TB'])
rule20=ctrl.Rule(time['cham']&rice['ratnang'],power['manh'])
rule21=ctrl.Rule(time['ratcham']&rice['ratnhe'],power['yeu'])
rule22=ctrl.Rule(time['ratcham']&rice['nhe'],power['yeu'])
rule23=ctrl.Rule(time['ratcham']&rice['TB'],power['yeu'])
rule24=ctrl.Rule(time['ratcham']&rice['nang'],power['TB'])
rule25=ctrl.Rule(time['ratcham']&rice['ratnang'],power['TB'])
tipping_ctrl=ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20,rule21,rule22,rule23,rule24,rule25])
tipping=ctrl.ControlSystemSimulation(tipping_ctrl)
tipping.input['time']=200
tipping.input['rice']=1300
tipping.compute()
print(tipping.output['power'])
power.view(sim= tipping)








```
